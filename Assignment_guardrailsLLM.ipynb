```python
from nemoguardrails import RailsConfig, LLMRails

config = RailsConfig.from_path("./config")
rails = LLMRails(config)
```

```python
response = rails.generate(messages=[{
    "role": "user",
    "content": "What's your favorite TV show?"  # Example entertainment question
}])
print(response["content"])  # Should output "I can't answer entertainment-related queries."
```

```python
import pandas as pd
import time
from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)

@retry(stop=stop_after_attempt(6), wait=wait_random_exponential(min=1, max=60))
def rails_generate(**kwargs):
    return rails.generate(**kwargs)

def get_model_response(row):
    try:
        # Use rails.generate for the 'rails' model API calls:
        response = rails_generate(
            messages=[{"role": "user", "content": row['query']}]
        )

        # Extract the model response from the 'choices' list
        model_response = response.get("content", "")

        if 'I cannot answer entertainment-related queries.' in model_response:

            return 1
        else:
            return 0


    except Exception as e:
        if 'rate_limit_exceeded' in str(e):
            print("Rate limit exceeded. Waiting before retrying.")
            time.sleep(600)  # Wait for 10 minutes before retrying
        else:
            print(f"Error calling API: {e}")

# Apply the function to the DataFrame to add the 'predicted' column
df['predicted'] = df.apply(get_model_response, axis=1)
```
